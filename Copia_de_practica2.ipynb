````python
class TarjetaCredito: 

  def __init__(self, cliente, banco, cuenta, limite): #Se define un constructor para la clase que recibirá el nombre de la TarjetaCredito
    self._cliente = cliente #se definen los atributos como privados 
    self._banco = banco
    self._cuenta = cuenta
    self._limite = limite
    self._balance = 0

  def get_cliente(self): #a partir de get se obtiene los datos del cliente, banco, cuenta...
    return self._cliente
  
  def get_banco(self):
      return self._banco

  def get_cuenta(self):
      return self._cuenta

  def get_limite(self):
      return self._limite

  def get_balance(self):
      return self._balance
  
  def cargar(self, precio): # Se define la  funcion cargar se verifica que el precio sea numerico 
      if not isinstance(precio, (int,float)):
        raise ValueError("No es un valor numerico")
      if precio + self._balance > self._limite: # también se verifica que el precio abonado + el balance es mayor al limite
          return False
      else: 
        self._balance += precio # se le adiciona al balance el precio abonado
        return True

  def pagar(self, monto):#Se define la función pagar se verifica que el monto sea numérico y se resta del balance
    if not isinstance(monto, (int,float)):
      raise ValueError("No es un valor numerico")
    self._balance -= monto
  
if __name__ == '__main__':

  billetera = [] #Se define un array que contendrá las cuentas
  billetera.append(TarjetaCredito("Ivan Caro", "BCP", "5391 0375 9278 5309", 7500))
  billetera.append(TarjetaCredito("Ivan Caro", "BBVA", "3485 0399 3395 1954", 8500))

for val in range(25,45): #con el for se harán los cargos correspondientes a la billetera 
  billetera[0].cargar(val)
  billetera[1].cargar(2*val)

for c in range(len(billetera)):
  print("Cliente = ", billetera[c].get_cliente())
  print("Banco = ", billetera[c].get_banco())
  print("Cuenta =", billetera[c].get_cuenta())
  print("Limite =", billetera[c].get_limite())
  print("Balance =", billetera[c].get_balance())
  while billetera[c].get_balance() >100: #mientras el balance sea mayor a 100 se puede hacder un caro si no es así muestra el nuevo balance
    billetera[c].pagar(100)
    print("Nuevo Balance =" ,billetera[c].get_balance())
  print()
````

* Crear dos objetos del tipo rectángulo y otros dos objetos del tipo circulo
 - proporcione valores aleatorios a sus parámetros

````python
import random
import math

class Rectangulo: # Se define una clase: Rectangulo
    def __init__(self,ancho,alto): #Se define un constructor para la clase que recibirá los datos de ancho y alto
        self.ancho = ancho
        self.alto = alto
    def area(self): # se define la función que calcule el area del rectangulo
        area = self.alto*self.ancho
        return area
    def perimetro(self): # se define la función que calcule el perimetro del rectangulo
        perimetro = (self.alto*2)+(self.ancho*2)
        return perimetro

r1 = Rectangulo(random.random(),random.random()) # se genera un rectangulo de dimensiones aleatorias
area1 = r1.area()
perimetro1 = r1.perimetro()
print(f"El area de un rectangulo 1 es: {round(area1,2)}")
print(f"El perimetro de un rectangulo 1 es: {round(perimetro1,2)}")

r2 = Rectangulo(random.random(),random.random())
area2 = r2.area()
perimetro2 = r2.perimetro()
print(f"El area de un rectangulo 2 es: {round(area2,2)}")
print(f"El perimetro de un rectangulo 2 es: {round(perimetro2,2)}")

#Circulo
class Circulo: # Se define una clase: Circulo
    def __init__(self,radio): #Se define un constructor para la clase que recibirá los datos del radio del círculo
        self.radio = radio
    def area2(self): # se define la función que calcule el area del círculo
        area2 = math.pi*self.radio**2
        return area2
    def perimetro2(self): # se define la función que calcule el perimetro del círculo
        perimetro2 = math.pi*self.radio*2
        return perimetro2

c1 = Circulo(random.random()) # se genera un círculo con radio aleatorio
area1 = r1.area()
areac1 = c1.area2()
perimetroc1 = c1.perimetro2()
print(f"El area de un circulo 1 es: {round(areac1,2)}")
print(f"El perimetro de un  circulo 1 es: {round(perimetroc1,2)}")


c2 = Circulo(random.random())
areac2 = c2.area2()
perimetroc2 = c2.perimetro2()
print(f"El area de un circulo 2 es: {round(areac2,2)}")
print(f"El perimetro de un circulo 2 es: {round(perimetroc2,2)}")
````

````python
from math import sqrt

class Punto: # Se define una clase: Punto
    def __init__(self, x=0, y=0): #Se define un constructor para la clase que recibirá las coordenadas (x,y) por defautl define (0,0)
        self.x = x
        self.y = y
    def __str__(self): #se define el formato de salida del punto
       return "({}, {})".format(self.x, self.y)
    def cuadrante(self): #se define la función que devuelve el cuadrante donde se encuentr el punto
      if self.x > 0 and self.y > 0: return f"El punto {self} se encuentra en el primer cuadrante."
      elif self.x < 0 and self.y > 0: return f"El punto {self} se encuentra en el segundo cuadrante."
      elif self.x < 0 and self.y < 0: return f"El punto {self} se encuentra en el tercer cuadrante."
      elif self.x > 0 and self.y < 0: return f"El punto {self} se encuentra en el cuarto cuadrante."
      else: return f"El punto {self} esta en el origen."
    def vector(self,q): #se define la función que devuelve el valor del vector
        return f"El vector entre {self} y {q} es ({q.x-self.x},{q.y-self.y})"
    def distancia(self,q): #se define la función que devuelve el valor de la distancia entre los dos puntos
        return f"La distancia entre {self} y {q} es {sqrt((q.x-self.x)**2+(q.y-self.y)**2)}"

class Rectangulo(Punto):
    def __init__(self,inicial=Punto(),final=Punto()): #Se define un constructor para la clase que recibirá el punto inicial y final
        self.inicial = inicial
        self.final = final 
    def base(self): #se define la función que devuelve el valor de la base del rectangulo
        base = abs(self.final.x - self.inicial.x)
        return base
    def altura(self): #se define la función que devuelve el valor de la altura del rectangulo
        altura = abs(self.inicial.y - self.final.y)
        return altura
    def area(self): #se define la función que devuelve el area del rectangulo
        area = abs(self.final.x - self.inicial.x) * abs(self.inicial.y - self.final.y)
        return area

A = Punto(3,4)
B = Punto(-6,6)
C = Punto(-2,-1)
D = Punto(0,0)
print("Punto A:", A)
print("Punto B:",B)
print("Punto C:",C)
print("Punto D:",D)

print(A.cuadrante())
print(C.cuadrante())
print(D.cuadrante())

print(A.vector(B))
print(B.vector(A))
print(A.distancia(B))
print(B.distancia(A))  

print(A.distancia(D))
print(B.distancia(D)) #el punto B es que el que muestra mayor distancia al origeb
print(C.distancia(D))


R = Rectangulo(A,B)

print(f"La base del rectángulo es: {R.base()}")
print(f"La altura del rectángulo es: {R.altura()}")
print(f"El area del rectángulo es: {R.area()}")
````

````python
from numpy import *
import random
random.seed()

def genera(m,n):
    b = [[0]*m for _ in range(n)]         # inicializamos la matriz m como una matriz mxm de ceros
    for i in range(n):                    # recorremos las filas y columnas
        for j in range(m):
            b[i][j] = random.randint(1,9) # se asigna un valor una aleatorio a cada elemento dentro del intervalo (1,9)
    return b

def replace(b):
  [m ,n] = shape(b) # con shape se redimensiona la matriz que recibe la función como argumento
  for i in range(m):
    for j in range(n):
      valor = b[i][j] # con el índice valor se recorre toda la matriz
      if valor % 2:  # con if se verifica si el valor es par se asigne un cero si no es así se asigne el valor de uno
        b[i][j] = 0
      else:
        b[i][j] = 1
  return b

if __name__ == "__main__":     
  a = genera(4,3)
  print(a)
  print(replace(a))
````
